# 1) Test if webcam works

```python
import cv2
import matplotlib.pyplot as plt

# Check if your webcam works and make a test snapshot
cap = cv2.VideoCapture(0)
ret, frame = cap.read()
if ret:
    print("Camera works!")
else:
    print("Camera does not work!")

# Convert to RGB
rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

plt.imshow(rgb_frame)
plt.axis('off')
plt.show()

cap.release()
```

# 2) Save snapshots

```python
import time
import os
from datetime import datetime

folder = "shared_artifacts/images/live_dataset"
os.makedirs(folder, exist_ok=True)

cap = cv2.VideoCapture(0)
print("Starting in 3 seconds, prepare your gesture!")
time.sleep(1)
print("Starting in 2 seconds, prepare your gesture!")
time.sleep(1)
print("Starting in 1 seconds, prepare your gesture!")
time.sleep(1)

num_images = 100 # how many photos to be taken
delay_frames = 0.75 # delay to not save frames constantly

for i in range(num_images):
    ret, frame = cap.read()
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S_%f")[:-3]
    filename = f"{folder}/img_{timestamp}.jpg"
    cv2.imwrite(filename, frame)
    print(f"Saved {i+1}/{num_images} â†’ {filename}")
    time.sleep(delay_frames)

cap.release()
print(f"Done {num_images} photos saved.")
```
