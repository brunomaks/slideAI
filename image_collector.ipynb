```python
import cv2
import matplotlib.pyplot as plt

# Check if your webcam works and make a test snapshot
cap = cv2.VideoCapture(0)
ret, frame = cap.read()
if ret:
    print("Camera works!")
else:
    print("Camera does not work!")

# Convert to RGB
rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

plt.imshow(rgb_frame)
plt.axis('off')
plt.show()

cap.release()
```

```python
import os
from datetime import datetime

folder = "shared_artifacts/images/live_dataset"
os.makedirs(folder, exist_ok=True)

cap = cv2.VideoCapture(0)
ret, frame = cap.read()

# Create a filename with current time
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S_%f")[:-3]
filename = f"{folder}/img_{timestamp}.jpg"

cv2.imwrite(filename, frame)
print(f"Saved: {filename}")

rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_frame)
plt.axis('off')
plt.title("Photo just saved!")
plt.show()

cap.release()
```
